## ¿Qué es Python?

Python es un lenguaje de programación de **alto nivel**, lo que significa que su sintaxis es muy parecida al lenguaje humano (específicamente al inglés). Se diseñó bajo la filosofía de que "el código se lee más veces de las que se escribe", por lo que es ideal para ingenieros que no son programadores de profesión pero necesitan herramientas eficaces.

## 3 Ejemplos sencillos y prácticos

Aquí tienes tres scripts básicos que ilustran cómo Python facilita tareas diarias en la agroindustria:

### 1. Monitoreo de Humedad de Granos

Un programa simple para decidir si un lote de grano está listo para almacenamiento seguro.

```
humedad = 13.5  # Porcentaje medido por el sensor

if humedad <= 14.0:
    print("Estado: Apto para almacenamiento.")
else:
    # Calculamos cuánto exceso hay
    exceso = humedad - 14.0
    print(f"Alerta: Humedad alta. Reducir {exceso}% antes de ensilar.")
```

### 2. Cálculo del Índice de Madurez

En la industria de jugos, la relación entre Sólidos Solubles (°Brix) y la Acidez es clave para el sabor.

```
brix = 12.8
acidez = 0.85

indice_madurez = brix / acidez

print(f"El índice de madurez es: {round(indice_madurez, 2)}")

# Criterio hipotético para cosecha de cítricos
if indice_madurez >= 12:
    print("Resultado: La fruta cumple con el estándar de dulzura.")
else:
    print("Resultado: La fruta aún está muy ácida.")
```

### 3. Rendimiento de Extracción de Aceite

Ideal para plantas procesadoras de oleaginosas o aceites esenciales.

```
def calcular_rendimiento(materia_prima, producto_final):
    rendimiento = (producto_final / materia_prima) * 100
    return rendimiento

# Ejemplo: 500kg de semilla producen 180kg de aceite
r = calcular_rendimiento(500, 180)

print(f"Rendimiento del proceso: {r}%")
```
